```python
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib inline

import cv2
import pytesseract

pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'
```

```python
carplate_img = [cv2.imread('images/IMG_3811.JPG'), cv2.imread('images/IMG_3812.JPG'), cv2.imread('images/IMG_5090.JPG')]
carplate_img_rgb = []
for img in carplate_img:
    carplate_img_rgb.append(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.imshow(carplate_img_rgb[0])
```

```python
carplate_haar_cascade = cv2.CascadeClassifier('haarcascade_russian_plate_number.xml')
```

```python
def carplate_detect(image):
    minNeighbors = 3
    carplate_overlay = image.copy() 
    carplate_rects = carplate_haar_cascade.detectMultiScale(carplate_overlay,scaleFactor=1.5, minNeighbors=minNeighbors)
    while len(carplate_rects) > 1:
        minNeighbors += 1
        carplate_rects = carplate_haar_cascade.detectMultiScale(carplate_overlay,scaleFactor=1.5, minNeighbors=minNeighbors)
    for x,y,w,h in carplate_rects: 
        cv2.rectangle(carplate_overlay, (x,y), (x+w,y+h), (255,0,0), 5) 
    return carplate_overlay
```

```python
plt.imshow(carplate_detect(carplate_img_rgb[0]))
```

```python
def carplate_extract(image):
    minNeighbors = 3
    carplate_rects = carplate_haar_cascade.detectMultiScale(image,scaleFactor=1.5, minNeighbors=minNeighbors)
    while len(carplate_rects) > 1:
        minNeighbors += 1
        carplate_rects = carplate_haar_cascade.detectMultiScale(image,scaleFactor=1.5, minNeighbors=minNeighbors)
    for x,y,w,h in carplate_rects: 
        carplate_img = image[y+15:y+h-10 ,x+15:x+w-20] 
    return carplate_img
```

```python
def enlarge_img(image, scale_percent):
    width = int(image.shape[1] * scale_percent / 100)
    height = int(image.shape[0] * scale_percent / 100)
    dim = (width, height)
    resized_image = cv2.resize(image, dim, interpolation = cv2.INTER_AREA)
    return resized_image
```

```python
carplate_extract_img = []
for i in carplate_img_rgb:
    carplate_extract_img.append(enlarge_img(carplate_extract(i), 150))
plt.imshow(carplate_extract_img[0]);
```

```python
carplate_extract_img_gray = []
for i in carplate_extract_img:
    carplate_extract_img_gray.append(cv2.cvtColor(i, cv2.COLOR_RGB2GRAY))
plt.axis('off') 
plt.imshow(carplate_extract_img_gray[0], cmap = 'Greys_r');
```

```python
carplate_extract_img_gray_blur = []
for i in carplate_extract_img_gray:
    carplate_extract_img_gray_blur.append(cv2.medianBlur(i,3))
plt.axis('off') 
plt.imshow(carplate_extract_img_gray_blur[0], cmap = 'Greys_r');
```

```python
c = 1
for i in carplate_extract_img_gray_blur:
    number = pytesseract.image_to_string(i, config = f'--psm 8 --oem 3 -c tessedit_char_whitelist=ABEKMHOPCTYX0123456789')
    if 7 < len(number) < 11:
        letters = 'ABEKMHOPCTYX'
        numbers = '0123456789'
        counter = 0
        for i in number:
            if (counter == 0 or counter == 4 or counter == 5) and i not in letters:
                number = number[:counter] + 'O' + number[counter+1:]
            if (counter == 1 or counter == 2 or counter == 3) and i not in numbers:
                number = number[:counter] + '0' + number[counter+1:]
            counter += 1
    print('(Фото №{0}) Номер после обработки: {1}'.format(c, number))
    c += 1
```
